# How to Stream Events
This notebook covers how to stream events from your graph (`stream_mode="events"`). Depending on the use case and user experience of your LangGraph application, your application may process event types differently.

```python
from langgraph_sdk import get_client

client = get_client()
```

```python
# create thread
thread = await client.threads.create()
thread
```

Streaming events produces responses containing an `event` key (in addition to other keys such as `data`). See the LangChain [`Runnable.astream_events()` reference](https://api.python.langchain.com/en/latest/runnables/langchain_core.runnables.base.Runnable.html#langchain_core.runnables.base.Runnable.astream_events) for all event types.

```python
# create input
input = {
    "messages": [
        {
            "role": "human",
            "content": "What's the weather in SF?",
        }
    ]
}

# stream events
async for chunk in client.runs.stream(
    thread_id=thread["thread_id"],
    assistant_id="agent",
    input=input,
    stream_mode="events",
):
    print(f"Receiving new event of type: {chunk.event}...")
    print(chunk.data)
    print("\n\n")
```
